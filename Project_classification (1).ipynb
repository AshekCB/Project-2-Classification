**Naive Bayes Classifier**

1.Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

2.Importing dataset


```python
df=pd.read_csv("/content/Iris.csv")
```

4.Feature Selection

```python
X=df.drop(['Species','Id'],axis=1)
Y=df['Species']
```

5. Importing Model and Training and testing:

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3)
```

```python
#fitting the data to the model
gnb.fit(x_train,y_train)
```

```python
y_pred=gnb.predict(x_test)
#predicting for the test data
```

6.Visulization

```python
sns.pairplot(df)#visualization
```

```python
sns.countplot(df)
#to check class imbalence problem
```

7.Predicting for samples

```python
gnb.predict(X.sample(n=5))
```

8.Model Evaluation

```python
from sklearn.metrics import accuracy_score,classification_report
report=classification_report(y_pred,y_test)
print("The Classification report is =\n\n",report)
score=accuracy_score(y_test,y_pred)
print("The Acuuracy Score is =\n",score*100)
```
